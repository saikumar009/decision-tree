```python
import numpy as np
import pandas as pd

data["fixed acidity"]=np.where(data["fixed acidity"]>9.20,9.20,data["fixed acidity"])
data["volatile acidity"]=np.where(data["volatile acidity"]>0.64,0.65,data["volatile acidity"])
data["citric acid"]=np.where(data["citric acid"]>0.42,0.42,data["citric acid"])
data["residual sugar"]=np.where(data["residual sugar"]>2.60,2.60,data["residual sugar"])
data["chlorides"]=np.where(data["chlorides"]>0.09,0.09,data["chlorides"])
data["free sulfur dioxide"]=np.where(data["free sulfur dioxide"]>21,21,data["free sulfur dioxide"])
data["total sulfur dioxide"]=np.where(data["total sulfur dioxide"]>62,62,data["total sulfur dioxide"])
data["sulphates"]=np.where(data["sulphates"]>0.73,0.73,data["sulphates"])
data["alcohol"]=np.where(data["alcohol"]>11.1,11.1,data["alcohol"])
data["quality"]=np.where(data["quality"]>6,6,data["quality"])
data["quality"]=np.where(data["quality"]<4,4,data["quality"])
```

```python
x=data.drop(columns=["quality"])
y=data['quality']
```

```python
scalar=StandardScaler()
x_scaled=scalar.fit_transform(x)
```

```python
x_scaled
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=35)
```

```python
dc=DecisionTreeClassifier()
```

```python
dc.fit(x_train,y_train)
```

```python
y_pred=dc.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
